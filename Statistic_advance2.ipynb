Question 4:  What are Type I and Type II errors? Give examples of each.

Ans>  
Type I error:

* Type I error, also known as a false positive, occurs in statistical hypothesis testing when a null hypothesis that is actually true is rejected. It's the error of incorrectly concluding that there is a significant effect or difference when there isn't one in reality.
* Example: Medical Testing: Suppose a medical test is designed to diagnose a particular disease. The null hypothesis (H0) is that the person does not have the disease, and the alternative hypothesis (H1) is that the person does have the disease. A Type I error occurs if the test incorrectly indicates that a person has the disease (rejects the null hypothesis) when they do not actually have it.

Type II error:

* Type II error, also known as a false negative, occurs in statistical hypothesis testing when a null hypothesis that is actually false is not rejected. In other words, it's the error of failing to detect a significant effect or difference when one exists in reality.
* Example: Medical Testing: In a medical test designed to diagnose a disease, a Type II error occurs if the test incorrectly indicates that a person does not have the disease (fails to reject the null hypothesis) when they actually do have it.

```python
# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.

import scipy.stats as stats
from statsmodels.stats.weightstats import ztest
import numpy as np
import pandas as pd


sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Assume population mean and standard deviation for the Z-test

population_mean =np.mean(sample_data)
print(f"Population Mean:",population_mean)
population_std_dev = np.std(sample_data)
print(f"Population Standard Deviation:",population_std_dev)

# Perform Z-test
z_statistic, p_value = ztest(sample_data, value=population_mean)

# Interpret the results
alpha = 0.05  # Significance level

print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Interpretation: There is enough evidence to suggest that the sample mean is significantly different from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Interpretation: There is not enough evidence to suggest that the sample mean is significantly different from the population mean.")
```

```python
# Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from statsmodels.stats.weightstats import zconfint

# Set parameters
mean = 0
std_dev = 1
sample_size = 1000

# Simulate data from a normal distribution
data = np.random.normal(mean, std_dev, sample_size)

# Calculate the 95% confidence interval for the mean
# We can use zconfint since we are simulating data from a known normal distribution
confidence_interval = zconfint(data, alpha=0.05)

# Plot the data using a histogram
plt.hist(data, bins=20, density=True, alpha=0.6, color='maroon')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Simulated Normal Distribution Data')
plt.show()

# Print the confidence interval
print(f"95% Confidence Interval for the Mean: {confidence_interval}")
```

```python
# Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent
# in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def calculate_z_scores(data):
  mean = np.mean(data)
  std_dev = np.std(data)
  z_scores = (data - mean) / std_dev
  return z_scores

# Generate some sample data
sample_data = np.random.normal(loc=50, scale=10, size=1000)

# Calculate Z-scores
z_scores = calculate_z_scores(sample_data)


# Visualize the standardized data (Z-scores) using a histogram
plt.hist(z_scores, bins=20, density=True, alpha=0.6, color='navy')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.title('Histogram of Z-scores (Standardized Data)')
plt.show()

# Explain what Z-scores represent
print("Explanation of Z-scores:")
print("A Z-score measures how many standard deviations away from the mean a particular data point is.")
print("A Z-score of 0 means the data point is equal to the mean.")
print("A positive Z-score means the data point is above the mean.")
print("A negative Z-score means the data point is below the mean.")
print("For example, a Z-score of 1 means the data point is 1 standard deviation above the mean, and a Z-score of -1.5 means the data point is 1.5 standard deviations below the mean.")
```
